```python
# Imports

import datetime
import pandas as pd
```

```python
# Import Data

df_fitnotes = pd.read_csv(r'..\test_data\FitNotes_Test_Data.csv')
df_myfitnesspal = pd.read_csv(r'..\test_data\MyFitnessPal_Test_Data.csv')
df_renpho = pd.read_csv(r'..\test_data\RenPho_Test_Data.csv')
```

```python
df_fitnotes['Volume (lbs)'] = df_fitnotes['Weight (lbs)'] * df_fitnotes['Reps']

df_fitnotes.head()
```

```python
dates = []

for date in df_renpho['Time of Measurement']:
    dates.append(datetime.datetime.strptime(date, "%B %d, %Y %I:%M:%S %p").date())

df_renpho['Date'] = dates

df_renpho.head()
```

```python
myfitnesspal_varsofinterest = ['Date', 'Calories', 'Fat (g)', 'Carbohydrates (g)', 'Protein (g)']

df_myfitnesspal_trimmed = df_myfitnesspal.filter(myfitnesspal_varsofinterest, axis=1)

df_myfitnesspal_trimmed.head()
```

```python
renpho_varsofinterest = ['Date', 'Weight(lb)', 'BMI', 'Body Fat(%)', 'BMR(kcal)', 'Metabolic Age']

df_renpho_trimmed = df_renpho.filter(renpho_varsofinterest, axis=1)

df_renpho_trimmed.head()
```
