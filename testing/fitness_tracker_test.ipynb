```python
# Imports

import datetime
import pandas as pd
```

```python
# Import Data

df_fitnotes = pd.read_csv(r'..\test_data\FitNotes_Test_Data.csv')
df_myfitnesspal = pd.read_csv(r'..\test_data\MyFitnessPal_Test_Data.csv')
df_renpho = pd.read_csv(r'..\test_data\RenPho_Test_Data.csv')
```

Now that we have imported the three datasets, let's do some data cleaning to get everything set up for analysis.

```python
df_fitnotes['Volume (lbs)'] = df_fitnotes['Weight (lbs)'] * df_fitnotes['Reps']

df_fitnotes.head()
```

```python
dates = []

for date in df_renpho['Time of Measurement']:
    dates.append(datetime.datetime.strptime(date, "%B %d, %Y %I:%M:%S %p").date())

df_renpho['Date'] = dates

df_renpho.head()
```

The RenPho dataset has one entry for each date since I weigh in once a day. The MyFitnessPal and FitNotes datasets have multiple entries per day. The MyFitnessPal has an entry for each meal (multiple meals a day) and the FitNotes dataset has an entry for each set performed of a workout (multiple sets per workout). My thought process is that since I am concerned with the sum totals of volume lifted and calories/macronutrients consumed per day, I can group the MyFitnessPal and FitNotes datasets by date, and then sum the needed values before combining the three datasets into one complete dataset.

```python
df_fitnotes_groupby_date = df_fitnotes.groupby('Date').sum()

df_fitnotes_groupby_date = df_fitnotes_groupby_date.reset_index()

df_fitnotes_groupby_date.head()
```

```python
df_fitnotes_groupby_date.info()
```

```python
df_myfitnesspal_groupby_date = df_myfitnesspal.groupby("Date").sum()

df_myfitnesspal_groupby_date = df_myfitnesspal_groupby_date.reset_index()

df_myfitnesspal_groupby_date.head()
```

```python
df_myfitnesspal_groupby_date.info()
```

```python
myfitnesspal_varsofinterest = ['Date', 'Calories', 'Fat (g)', 'Carbohydrates (g)', 'Protein (g)']

df_myfitnesspal_grouped_trimmed = df_myfitnesspal_groupby_date.filter(myfitnesspal_varsofinterest, axis=1)

df_myfitnesspal_grouped_trimmed.head()
```

```python
df_renpho.info()
```

```python
renpho_varsofinterest = ['Date', 'Weight(lb)', 'BMI', 'Body Fat(%)', 'BMR(kcal)', 'Metabolic Age']

df_renpho_trimmed = df_renpho.filter(renpho_varsofinterest, axis=1)

df_renpho_trimmed.head()
```

The RenPho data set it the only one that shows the Date column differently that the other two datasets. I am going to try to reverse the order of the Dates in df_renpho to see if that makes the data easier to merge.


```python
df_renpho_trimmed_reversed = df_renpho_trimmed.iloc[::-1]

df_renpho_trimmed_reversed = df_renpho_trimmed_reversed.reset_index().drop('index', axis=1)

df_renpho_trimmed_reversed.head()
```

Now it is time to combine these three dataframes from the three different apps into one complete data source.

```python
df_combined = pd.concat([df_myfitnesspal_grouped_trimmed, df_fitnotes_groupby_date, df_renpho_trimmed_reversed], axis=1)

df_combined.head()
```
