```python
# Imports

import pandas as pd 
import numpy as np
```

```python
# Import data

df = pd.read_csv('data\FitNotes\FitNotes_Export_2021_11_30_15_21_26.csv')

df.head()
```

```python
# Which exercise has been peformed most often (sets)?

frequency = df["Exercise"].value_counts()
frequency.head()
```

```python
# Which exercise has been performed most often (reps)?

exercises_grouped = df.groupby('Exercise')

exercises_grouped['Reps'].sum().sort_values(ascending=False).head()
```

```python
# Which exercise was I doing when I lifted the most weight?

df.sort_values(by='Weight (lbs)', ascending=False).head()
```

```python
# Which exercise have I used to lift the most volume?

df['Volume'] = df['Weight (lbs)'] * df['Reps']

exercises_grouped = df.groupby('Exercise')
exercises_grouped.sum().sort_values(by='Volume', ascending=False).head()
```
